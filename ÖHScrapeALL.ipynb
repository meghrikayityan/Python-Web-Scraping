# Extracting "Data" from a house-rentals website in Salzburg
# web scraping
-> extracting data from 15 pages 

result-> 149 rows(active listings) / 6 colums


P.S.
TypeWG = "WohnGemeinschaft" = Sharing apartments
 // TypeEntire= Entire apartments

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup


L = []
baseURL = "https://boersen.oeh-salzburg.at/boersen/wohnungsboerse/page/1/?amount=‚Ç¨%200%20-%20‚Ç¨%201000&cp_art_des_angebots%5B0%5D=Angebot%20%28Biete%29&sa=Go&refine_search=yes"
for pages in range(1,16):
    print("https://boersen.oeh-salzburg.at/boersen/wohnungsboerse/page/" + str(pages) + "/?amount=‚Ç¨%200%20-%20‚Ç¨%201000&cp_art_des_angebots%5B0%5D=Angebot%20%28Biete%29&sa=Go&refine_search=yes")
    r = requests.get("https://boersen.oeh-salzburg.at/boersen/wohnungsboerse/page/" + str(pages) + "/?amount=‚Ç¨%200%20-%20‚Ç¨%201000&cp_art_des_angebots%5B0%5D=Angebot%20%28Biete%29&sa=Go&refine_search=yes")
    c = r.content
    soup = BeautifulSoup(c, "html.parser")
    x = soup.find_all("div", {"class": "post-block"})
    for i in x:
        dic = {}

        z = i.find("span", {"class": "dashicons-before owner"})
        dic["Publisher"] = z.find("a").text


        if "WG" in i.find("h3").text:
            dic["TypeWG"] = "yes"
        elif "Wg" in i.find("h3").text:
            dic["TypeWG"] = "yes"
        else:
            dic["TypeEntire"] = "yes"

        dic["Price"] = i.find("p", {"class": "post-price"}).text

        dic["Description"] = i.find("p", {"class": "post-desc"}).text.replace("[‚Ä¶]","")
        y = i.find_all("h3")
        for j in y:
            for a in j.find_all("a", href=True):
                dic["Link"] = a['href']

        L.append(dic)
```

```python
dataframe = pd.DataFrame(data = L)

dataframe

dataframe.to_excel("√ñHWohnb√∂rsenALL.xlsx")
```
